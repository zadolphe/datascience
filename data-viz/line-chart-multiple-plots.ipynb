<h2>You can add multiple plots to a subplot or multiple plots to the base subplot which is the initialized subplot when creating the figure class, see below:

<h3> We will also use the datetime module, to extract the month values from the DATE column and assign them to a new column, we can use the pandas.Series.dt accessor:
    unrate['MONTH'] = unrate['DATE'].dt.month

```python
unrate['MONTH'] = unrate['DATE'].dt.month
fig = plt.figure(figsize=(6,3))
plt.plot(unrate[0:12]['MONTH'],unrate[0:12]['UNRATE'], c='red')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['UNRATE'], c='blue')

plt.show()
```

```python
#use a loop to plot multiple plots on a base subplot with different colors:
fig = plt.figure(figsize=(10,6))
colors = ['red','blue','green','orange','black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    date_range = unrate[start_index:end_index]
    plt.plot(date_range['MONTH'],date_range["UNRATE"],c=colors[i])
    
plt.show()
```

```python
#add legend and label
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = 1948 + i
    plt.plot(subset['MONTH'], subset['UNRATE'], c=colors[i], label=label)

plt.legend(loc='upper left')
plt.show()
```

<h2>Bar plots and scatter plots 

```python
from numpy import arange
import pandas as pd
```

```python
reviews = pd.read_csv('fandango_score_comparison.csv')
reviews.head(3)
```

```python
norm_reviews = pd.DataFrame(reviews[['FILM','RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']])
norm_reviews.iloc[[0],:]
##print(norm_reviews[0:2])
```

```python
#create a bars from a value of a row in reviews onto a figure and axes, declare fig and axes using shortcut with plt.subplots
#axes.bar takes list like values for x (position of bars) and y(heights of the bars) and a scalar value for width
#use numpy arange to create a list of values which become the bar positions and start the first bar with its center at 0.75, they will be evenly spaced
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_heights = norm_reviews[num_cols].iloc[0].values
bar_positions = arange(5) + 0.75

fig, ax = plt.subplots()
ax.bar(bar_positions,bar_heights, 0.5)
```

```python
#set the ticks on the x axis to their positions, set xticks' labels and rotate them 90 degrees, and set x-axis label. Set the title of the y-axes and axes
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_heights = norm_reviews[num_cols].iloc[0].values
bar_positions = arange(5) + 0.75
tick_positions = range(1,6)

fig, ax = plt.subplots()
ax.bar(bar_positions,bar_heights, 0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols, rotation=90)
ax.set_xlabel("Rating Source")

ax.set_ylabel("Average Rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")

plt.show()
```

<h2>Horizontal bar plot

```python
#same approach and methods just for axes.barh for horizontal bar plot
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_widths = norm_reviews[num_cols].iloc[0].values
bar_positions = arange(5) + 0.75
tick_positions = range(1,6)

fig, ax = plt.subplots()
ax.barh(bar_positions,bar_widths,0.5)
ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols)
ax.set_ylabel("Rating Source")
ax.set_xlabel("Average Rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron (2015)")
plt.show()
```

<h2>Histograms and Boxplots

```python
reviews = pd.read_csv('fandango_score_comparison.csv')
cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']
norm_reviews = reviews[cols]
print(norm_reviews[:5])
```

```python
fandango_distribution = norm_reviews['Fandango_Ratingvalue'].value_counts().sort_index()

imdb_distribution = norm_reviews['IMDB_norm'].value_counts().sort_index()

print(fandango_distribution)
print(imdb_distribution)
```

```python
#simple frequency distribution of fandango ratings ranging from 0 to 5
fig, ax = plt.subplots()
ax.hist(norm_reviews['Fandango_Ratingvalue'], range=(0,5))
plt.show()
```

We can visually examine the proportional area that the bars in the 2.0 to 4.0 range take up and determine that more than 50% of the movies on Fandango fall in this range. We can increase the number of bins from 10 to 20 for improved resolution of the distribution. The length of each bin will be 0.25 (5 / 20) instead of 0.5 (5 / 10). The bins parameter for Axes.hist() is the 2nd positional parameter, but can also be specified as a named parameter:

 To ensure that the scales for the y-axis are the same for all histograms, let's set them manually using Axes.set_ylim().

```python
#generate histograms with different number of bins and ranges 
fig = plt.figure(figsize=(5,20))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)

ax1.hist(norm_reviews['Fandango_Ratingvalue'],bins=20,range=(0,5))
ax2.hist(norm_reviews['RT_user_norm'],bins=20,range=(0,5))
ax3.hist(norm_reviews['Metacritic_user_nom'],bins=20,range=(0,5))
ax4.hist(norm_reviews['IMDB_norm'],bins=20,range=(0,5))

ax1.set_ylim(0,50)
ax2.set_ylim(0,50)
ax3.set_ylim(0,50)
ax4.set_ylim(0,50)

ax1.set_title("Distribution of Fandango Ratings")
ax2.set_title("Distribution of Rotten Tomatoes Ratings")
ax3.set_title("Distribution of Metacritic Ratings")
ax4.set_title("Distribution of IMDB Ratings")

plt.show()
```
