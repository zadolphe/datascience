```python
#define functions for changing numeric types from being stored as string objects
def convert_price(price):
    price = price.strip()
    price = price.replace("$","")
    price = price.replace(",","")
    return price
def clean_odometer(odometer):
    odometer = odometer.strip()
    odometer = odometer.replace(",","")
    odometer = odometer.replace("km","")
    return odometer
```

```python
numeric_price = []
numeric_odometer = []

for p in autos['price']:
    p = convert_price(p)
    numeric_price.append(p)
    
for o in autos['odometer']:
    o = clean_odometer(o)
    numeric_odometer.append(o)
    
autos['price'] = numeric_price
autos['odometer'] = numeric_odometer
    
autos['price'] = autos['price'].astype('int')
autos['odometer'] = autos['odometer'].astype('int')
```
